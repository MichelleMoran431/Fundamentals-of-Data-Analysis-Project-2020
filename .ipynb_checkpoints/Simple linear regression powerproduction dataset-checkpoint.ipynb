3. Splitting the dataset

The dataset is split into test and train set, following a 20-80 or 30-70 rule. It is necessary to split in order to train our model according to the wind speed and power generated. Then we test the model dataset on the test set. 

Checking if the predictions made by the model on the testset data matches wehat is given in the dataset.  If it matches , it imples that out model is accurate and is making the right predictions.

```python
#Splitting the dataset

x = dfpower.speed
y = dfpower.power

from sklearn.model_selection import train_test_split
import sklearn.model_selection as model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.3, random_state = 0)
print ("x_train: ", x_train)
print ("y_train: ", y_train)
print("x_test: ", x_test)
print ("y_test: ", y_test)
```

Using SKlearns linear model library , linear regression class is imported  and to create an object for a linear regression class called regressor. Then using the fit method , to fit the regressor into the training set, fitting the x_train values into the target y_train values. Thus the model learns the correlation and predictions on the dependt vairbales based on the independent variables can be done.

```python
#Fitting linear regression model into the training set
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd


x = dfpower.speed
y = dfpower.power 

x_train = x.values.reshape(-1, 1)
y_train = y.values.reshape(-1, 1)

x_test = x.values.reshape(-1, 1)
y_test = y.values.reshape(-1, 1)

regressor = LinearRegression()
regressor.fit(x_train,y_train) #actually produces the linear eqn for the data
```

5. Predicting the test set results

A vector is created containing all the predictions of the test set power values. The predicted power values are then put into the vector called y_pred( containing prediction for all observations in the test set)

Predict method makes the predictions for the test set.

y_test = real power values for the test set
y_pred = are the predicted power values 

```python
#Predicting the test set results

y_pred = regressor.predict(x_test) 
y_pred
```

##### Visualizing the results

To visualize the results we plot the graphs using matplotlib.
plt.scatter plots a scatter plot of the data : 

 - X – coordinate (X_train: Windspeed)
 - Y – coordinate (y_train: Power)
 - Color ( Regression line in red and observation line in blue)


```python
#Visualizing the results
#plot for the TRAIN
import matplotlib.pyplot as plt 
plt.scatter(x_train, y_train, color='red') # plotting the observation line
plt.plot(x_train, regressor.predict(x_train), color='blue') # plotting the regression line
plt.title("Windspeed vs Power (Training set)") # stating the title of the graph
  
plt.xlabel("Wind speed") # adding the name of x-axis
plt.ylabel("Power") # adding the name of y-axis
plt.show() # specifies end of graph
 
#plot for the TEST
  
plt.scatter(x_test, y_test, color='red') 
plt.plot(x_train, regressor.predict(x_train), color='blue') # plotting the regression line
plt.title("Windspeed vs Power (Testing set)")
  
plt.xlabel("Wind speed") 
plt.ylabel("Power") 
plt.show() 
```

```python
from sklearn.metrics import mean_squared_error, r2_score


# The coefficients
print('Coefficients: \n', regressor.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

x = dfpower.speed
y = dfpower.power 

x_train = x.values.reshape(-1, 1)
y_train = y.values.reshape(-1, 1)

x_test = x.values.reshape(-1, 1)
y_test = y.values.reshape(-1, 1)





# Split the data into training/testing sets
x_train = x_train[:-20]
x_test = x_test[-20:]

# Split the targets into training/testing sets
y_train = y_train[:-20]
y_test = y_test[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(x_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test,y_pred))

# Plot outputs
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
